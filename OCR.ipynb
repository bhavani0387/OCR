#### View Image

```python
# Import the necessary libraries  
import time 
import sys
import cv2
import pytesseract 
from PIL import Image
from skimage import io
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

### Different ways to open & Display the image

```python
# Open the image file 
img = Image.open('images/Provisional-Hospital-Bill.png')
 
img.show()

time.sleep(5)
```

```python
img = cv2.imread('images/Provisional-Hospital-Bill.png')
while True:
    cv2.imshow("Provisional Hospital Bill", img)
    cv2.waitKey(0)
    sys.exit() # to exit from all the processes
    #The image will open in another window.
 
cv2.destroyAllWindows() # Click esc to close the image.
```

```python

img = mpimg.imread('images/Provisional-Hospital-Bill.png')

# Display the image with different interpolation methods
plt.figure(figsize=(10, 10))

# Original image
plt.subplot(2, 2, 1)
plt.imshow(img)
plt.title('Original')
plt.axis('off')

# Nearest interpolation
plt.subplot(2, 2, 2)
plt.imshow(img, interpolation='nearest')
plt.title('Nearest')
plt.axis('off')

# Bilinear interpolation
plt.subplot(2, 2, 3)
plt.imshow(img, interpolation='bilinear')
plt.title('Bilinear')
plt.axis('off')

# Bicubic interpolation
plt.subplot(2, 2, 4)
plt.imshow(img, interpolation='bicubic')
plt.title('Bicubic')
plt.axis('off')

plt.show()
```
